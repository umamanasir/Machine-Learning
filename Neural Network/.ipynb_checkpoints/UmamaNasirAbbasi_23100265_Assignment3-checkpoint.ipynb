### Task 2.2: Tensorflow Keras

In this part you will use the [Keras](https://www.tensorflow.org/api_docs/python/tf/keras/Sequential) to implement the [Neural Network](https://machinelearningmastery.com/tutorial-first-neural-network-python-keras/) and apply it to the MNIST audio dataset (provided in part 1). Split the training dataset into train and evaluation data with 90:10 ratio. Run evaluation on X_eval while training on X_train. Tune the hyperparameters to get the best possible classification accuracy. You need to report accuracy, recall, precision and F1 score on the test dataset and print the confusion matrix.

````python
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Input,Dense
from keras import optimizers
````

````python
#split into test and validation dataset. 

Xtrain, x_eval, Ytrain, y_eval = train_test_split(X_train, Y_train, test_size=0.10,shuffle = True)  
print(Xtrain.shape)
print(Ytrain.shape)
print(x_eval.shape)
print(y_eval.shape)
````

````python
# Set the parameters accordingly
LEARNING_RATE = 0.001
BATCH_SIZE = 100
EPOCHS = 50
SHAPE = (13,)
````

````python
# neural network.
model = Sequential()
model.add(Dense(300, input_shape = SHAPE, activation = 'relu')) #hidden layers.
model.add(Dense(150, activation = 'relu')) 
model.add(Dense(100, activation = 'relu')) 
model.add(Dense(80, activation = 'softmax')) #output layer. 

adam = optimizers.Adam(learning_rate=LEARNING_RATE)
model.compile(loss ="sparse_categorical_crossentropy" , optimizer = adam, metrics = ['accuracy']) #loss = categorical cross entopy. 
model.fit(x = Xtrain, y = Ytrain, epochs=EPOCHS, batch_size=BATCH_SIZE)
````

````python
_, accuracy = model.evaluate(x_eval, y_eval,verbose = 1) #testing on evaluation data. 
````

````python
predictions = np.argmax ( model.predict(X_test), axis =1) #testing on test daata. 
print(classification_report(Y_test, predictions))
print(confusion_matrix(Y_test, predictions))
````

In this part you will use the [Keras](https://pytorch.org/docs/stable/nn.html) to implement the [Neural Network](https://medium.com/analytics-vidhya/a-simple-neural-network-classifier-using-pytorch-from-scratch-7ebb477422d2) and apply it to the MNIST audio dataset (provided in part 1).
````python
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from sklearn.metrics import accuracy_score
````

````python
Xtrain, x_eval, Ytrain, y_eval = train_test_split(X_train, Y_train, test_size=0.10,shuffle = True) #split data set. 
print(Xtrain.shape)
print(Ytrain.shape)
print(x_eval.shape)
print(y_eval.shape)
````

````python
class Data(Dataset):
    def __init__(self, X_train, y_train):
        # Code here
        ...
        self.x = torch.from_numpy(X_train.astype(np.float32))
        self.y = torch.from_numpy(y_train).type(torch.LongTensor)
        self.len = self.x.shape[0]

    def __getitem__(self, index):
        # Code here
        return self.x[index], self.y[index]
    
    def __len__(self):
        # Code here
        ...
        return self.len
````

````python
class NeuralNetwork(nn.Module):
    def __init__(self):
        super(NeuralNetwork, self).__init__()
        self.linearin = nn.Linear(input_dim, hidden_layers)
        self.linearout = nn.Linear(hidden_layers, output_dim)
        # Code here

    def forward(self, x):
        # Code here
        x_hat = torch.relu(self.linearin(x))
        x = self.linearout(x_hat)
        return x
````

````python
# Set the parameters accordingly
LEARNING_RATE = 0.1
BATCH_SIZE = 50
EPOCHS = 70
hidden_layers = 150
output_dim = 80
input_dim = 13
````

````python
training = Data(np.asarray(Xtrain), np.asarray(Ytrain))
validation = Data(np.asarray(x_eval), np.asarray(y_eval))
````

````python
# Set the loss function and optimizer accordingly
model = NeuralNetwork()
loss_function = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=LEARNING_RATE)

trainLoader = DataLoader(training, BATCH_SIZE, shuffle = True, num_workers = 2)
validLoader = DataLoader(validation, BATCH_SIZE, shuffle = True, num_workers = 2)
````

````python
for epochs in range(EPOCHS):
  r_loss = 0.0
  val_accuracy = -.0
  val_loss = 0.0
  prediction = 0
  total_val = 0
  total_train = 0
  for data in trainLoader: 
    values, true =data
    optimizer.zero_grad()
    predict = model(values) #forward 
    loss = loss_function(predict, true)
    loss.backward() #backprop. 
    optimizer.step() #optimize
    r_loss = r_loss + loss.item()
    total_train = total_train + true.size(0)
    # print(predict.shape, values.shape ,true.shape)
    _, predicted = torch.max(predict, 1)
    prediction = prediction + (predicted==true).sum().item()
  loss_training = r_loss / len(trainLoader)
  accuracy = prediction / total_train
  with torch.no_grad():
    model.eval()
    for data in validLoader:
      values, true = data
      predict = model(values)
      loss_value = loss_function(predict, true)
      _, predicted = torch.max(predict, 1)
      r_loss = r_loss + loss_value.item()
      total_val = total_val + true.size(0)
      val_accuracy = val_accuracy + (predicted == true).sum().item()
  valid_loss = r_loss / len(validLoader)
  accuracy_valid = val_accuracy / total_val
  
  print("Epoch: ", epochs, '/', EPOCHS)
  print('Loss:', loss_training, ' - ', 'accuracy:', accuracy, ' - ', 'val_loss:', valid_loss, ' - ', 'val_accuracy:', accuracy_valid)
````

````python
testData = Data(np.asarray(X_test), np.asarray(Y_test))
testingLoader = DataLoader(testData, BATCH_SIZE, shuffle = True)
````

````python
correct, all = 0,0
prediction = []
labels = []
with torch.no_grad():
  for data in testingLoader: 
    value, true_labels = data
    predict = model(value)
    _, predicted = torch.max(predict, 1)
    correct = correct + (predicted == true_labels).sum().item()
    all = all + true_labels.size(0)
    prediction.append(list(predicted))
    labels.append(list(true_labels))

print("Accuracy on test data = ", (correct/all)*100 , '%')
````

````python
prediction = np.array(prediction).flatten()
labels = np.array(labels).flatten()
print(classification_report(labels, prediction))
print(confusion_matrix(labels, prediction))
````

**Q1**: Compare the tools you used above. State the advantages and disadvantages.

**Q2**: What is the purpose of the data loader in PyTorch?

**ANSWERS:**

Q1: sklearn highly abstracts the process of building a neural network. It allows users to create neural networks rapidly and even people who do not have a very good idea of the mechanics of neural networks such as hidden layers, activation functions, or learning rate (alpha) can build models using sklearn. Thus the main advantage is that it is very simple. However it is relatively slower than other tools. 

On the other hand, while keras also provide a high level of abtraction, it does give the user bit more control over how the model can be built. It allows user to specify activation functions in each layer, manually inputting layers into the model and working and tuning the layer to decide the layer sizes, activation functions etc. Although relatively less simple, it is very powerful and faster than sklearn. 

Lastly Pytorch allows a high level of customization in the neural network which is it's main advantage.
